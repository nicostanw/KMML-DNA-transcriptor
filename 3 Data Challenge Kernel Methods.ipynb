Creation of the Exponential Kernel

```python
K_exp=np.zeros((n,n))
#print(K_exp)
#print(K_exp.shape)
```

```python
for i in range(1,n+1):
    mat=X[:n-i,]-X[i:n,]
    square_norm=np.sqrt(np.sum(np.abs(mat)**2,axis=-1))
    np.fill_diagonal(K_exp[:-i,i:],np.exp(-square_norm))
    np.fill_diagonal(K_exp[i:,:-i],np.exp(-square_norm))

np.fill_diagonal(K_exp,np.ones(n))
print(K_exp)
```

Creation of the Gaussian Kernel (sigma_square=1)

```python
K_gaussian=np.zeros((n,n))
#print(K_gaussian)
#print(K_gaussian.shape)
```

```python
for i in range(1,n+1):
    mat=X[:n-i,]-X[i:n,]
    square_norm=np.sum(np.abs(mat)**2,axis=-1)
    np.fill_diagonal(K_gaussian[:-i,i:],np.exp(-square_norm))
    np.fill_diagonal(K_gaussian[i:,:-i],np.exp(-square_norm))

np.fill_diagonal(K_gaussian,np.ones(n))
print(K_gaussian)
```

# KLR

```python
def l(x):
    return np.log(1+np.exp(-x))
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def l_prime(x):
    return -sigmoid(-x)
```

```python
def l_second(x):
    return sigmoid(x)*sigmoid(-x)
```

```python
import scipy
from scipy.linalg import sqrtm
```

```python
def IRLS(K,alpha_prev):
    
    P=l_prime(Y*np.dot(K,alpha_prev))
    P=np.diag(P)
    
    W=l_second(Y*np.dot(K,alpha_prev))
    W=np.diag(W)
    W_sqr=scipy.linalg.sqrtm(W)
    
    alpha_0=alpha_prev+np.random.normal(0, 0.01, n)
    Z=np.dot(K,alpha_0)-np.dot(np.linalg.inv(W),np.dot(P,Y))
    
    alpha=np.dot(np.dot(np.dot(W_sqr,np.linalg.inv(np.dot(W_sqr,np.dot(K,W_sqr))+n*lbd*np.eye(n))),W_sqr),Z)
    
    return alpha
```

```python
def KLR(K,alpha,alpha_prev,eps,lbd=1):
    cpt=0
    while np.linalg.norm(alpha-alpha_prev)>=eps:
        alpha_prev=alpha
        alpha=IRLS(K,alpha_prev)

        if cpt%10==0:
            print(cpt)
        cpt+=1

    return(alpha)  
```

```python
from qpsolvers import solve_qp
```

```python
def SVM(K,lbd=1):
    P=K.astype(np.float)
    q=-Y.astype(np.float)
    G=-np.diag(Y).astype(np.float)
    h=np.zeros(n).astype(np.float)
    A=np.zeros((n,n)).astype(np.float)
    b=np.zeros(n).astype(np.float)
    lb=np.repeat(float('-inf'),n)
    ub=1/(2*lbd*n*Y).astype(np.float)
    
    alpha_star=solve_qp(P, q, G, h, A, b, lb, ub)
    return alpha_star
```

```python
import sympy as sp
from sympy.parsing.sympy_parser import parse_expr
from sympy.utilities.lambdify import lambdastr, lambdify
from sympy.solvers import solve

import numpy as np

import matplotlib.pyplot as plt
```

```python
def calc_conjugate(str, varnames='x'):

    # set the symbols
    vars = sp.symbols(varnames)
    x = vars[0] if isinstance(vars, tuple) else vars
    y = sp.symbols('y', real=True)

    # set the function and objective
    fun = parse_expr(str)
    obj = x*y - fun

    # calculate derivative of obj and solve for zero
    sol = solve(sp.diff(obj, x), x)

    # substitute solution into objective
    solfun = sp.simplify(obj.subs(x, sol[0]))

    # if extra values were passed add to lambda function
    varnames = [y] + list(vars[1:]) if isinstance(vars, tuple) else y

    return (sp.sstr(solfun), lambdify(vars, fun, 'numpy'), lambdify(varnames, solfun, 'numpy'))
```
